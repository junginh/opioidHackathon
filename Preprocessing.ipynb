```python
import json
import re
import pandas as pd
import os
from time import sleep
```

```python
def deEmojify(inputString):
    return inputString.encode('ascii', 'ignore').decode('ascii')
# https://stackoverflow.com/questions/33404752/removing-emojis-from-a-string-in-python
```

```python
def json_to_txt(inputfile:str,outputfile:str):
    with open(inputfile, "r", encoding="utf8") as jsoninput:
        data = jsoninput.read()
    obj = json.loads(data)
    f = open(outputfile, "w+")
    for i in range(10000):
        f.write(deEmojify(obj[i]["text"]))
    f.close()
```

```python
def txt_to_list(inputfile) -> list:
    f = open(inputfile, "r")
    wordList = []
    lines = f.readlines()

    for x in lines:
        #print(x)
        x = re.sub(r'@', ' @', x)
        x = re.sub(r'http[^ ]*', ' ', x)
        x = re.sub(r'[\(|\)]+', ' ', x)
        x = re.sub(r'@[\w]*', ' ', x)
        x = re.sub(r'[^A-Za-z0-9_\'-,]+', ' ', x)
        wordList.extend(x.lower().split())
        
    return wordList
```

```python
def count_word_frequency(wordList, vocab):
    totWords = len(wordList)
    frequency_dict = {}
    for i in vocab:
        frequency_dict[i] = 0

    for j in wordList:
        if j in frequency_dict.keys():
            frequency_dict[j] += 1
    for i in vocab:
        frequency_dict[i] = frequency_dict[i]/totWords
        
    return frequency_dict
```

```python
def process_files(file_list, vocab):
    matrix = {}
    dict_list = []
    for file in file_list:
        txt_file = file.split(".")[0]+".txt"
        print(file)
        json_to_txt(file, txt_file)
        print("{} to txt succeed".format(file))
        wordList = txt_to_list(txt_file)
        print("{} to list succeed".format(file))
        frequency_dict = count_word_frequency(wordList, vocab)
        print("{} to dict succeed".format(file))
        dict_list.append(frequency_dict)
    
    for word in vocab:
        matrix[word] = [dict_list[i][word] for i in range(len(dict_list))]
    
    time_list = []
    for file in file_list:
        time_list.append(file.split("/")[-1].split(".")[0])
    matrix['Time'] = time_list
    pd.DataFrame(matrix).to_csv('Matrix.csv')
```
